```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
# read ./elma-bridges.csv as geoDataFrame
allbridges = gpd.read_file('./data/all-bridges.csv')
allbridges['longitude'] = allbridges['GPS (Longitude)'].astype(float)
allbridges['latitude'] = allbridges['GPS (Latitude)'].astype(float)
# create geomtry column from latitude and longitude columns
allbridges['geometry'] = allbridges.apply(lambda x: Point((float(x.longitude), float(x.latitude))), axis=1)
allbridges = allbridges.drop(['longitude', 'latitude'], axis=1)
crs = {'init': 'epsg:4326'}
allbridges
```

```python
# return all row that matches 1017673 in the project code column
allbridges[allbridges['Project Code'] == "1017673"]
```

```python
# read ./elma-bridges.csv as geoDataFrame
elma = gpd.read_file('./data/elma-bridges.csv')
```

```python
# convert longitude and latitude to float
elma['longitude'] = elma['longitude'].astype(float)
```

```python
# create geomtry column from latitude and longitude columns
elma['geometry'] = elma.apply(lambda x: Point((float(x.longitude), float(x.latitude))), axis=1)
```

```python
# creating a geometry column 
# geometry = [Point(xy) for xy in zip(elma['longitude'], elma['latitude'])]
elma = elma.drop(['longitude', 'latitude'], axis=1)
crs = {'init': 'epsg:4326'}
```

```python
elma
```

```python
# write elma to geojson
elma.to_file('./data/elma-bridges.geojson', driver='GeoJSON')
```

```python
# create geodataframe from allbridges of rows that do  not have a value in the project code column that matches a value in the project code column of elma
diffbridges = allbridges[~allbridges['Project Code'].isin(elma['Project Code'])]
```
